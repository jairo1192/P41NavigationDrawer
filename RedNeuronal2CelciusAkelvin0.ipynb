```python
import tensorflow as tf
import numpy as np
```

```python
celsius = np.array([10, 24.15, 0, -12, 18, 54, 37, -175, 4, 76.85], dtype=float)
kelvin = np.array([283.15, 297.30, 273.15, 261.15, 291.15, 327.15, 310.15, 98.15, 277.15, 350], dtype=float)
```

```python
capa = tf.keras.layers.Dense(units=1, input_shape=[1])
modelo = tf.keras.Sequential([capa])
```

```python
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss = 'mean_squared_error'
)
```

```python
print("Entrenando Modelo...")
historial = modelo.fit(celsius, kelvin, epochs=3000, verbose=False)
print("Modelo Entrenado! :D")
```

```python
import matplotlib.pyplot as plt
plt.xlabel("# Ciclos")
plt.ylabel("Magnitud de perdida")
plt.plot(historial.history["loss"])
```

```python
print("Comenzando Pruebas")
resultado = modelo.predict([100.0])
print("El resultado es " + str(resultado) + " grados Kelvin")
```
